## Import Libraries
**Import pandas,seaborn, and the usual libraries.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the Data
** Read the 'KNN_Project_Data csv file into a dataframe **

```python
df = pd.read_csv('KNN_Project_Data.csv')
```

**Check the head of the dataframe.**

```python
df.head()
```

# EDA

Since this data is artificial, we'll just do a large pairplot with seaborn.

**Use seaborn on the dataframe to create a pairplot with the hue indicated by the TARGET CLASS column.**

```python
sns.set_style('darkgrid')
sns.pairplot(df,hue='TARGET CLASS', palette ='bwr', diag_kind='hist')
```

# Standardize the Variables

Time to standardize the variables.

**Import StandardScaler from Scikit learn.**

```python
from sklearn.preprocessing import StandardScaler
```

**Create a StandardScaler() object called scaler.**

```python
scaler = StandardScaler()
```

**Fit scaler to the features.**

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

**Use the .transform() method to transform the features to a scaled version.**

```python
sfeatures = scaler.transform(df.drop('TARGET CLASS', axis =1))
```

**Convert the scaled features to a dataframe and check the head of this dataframe to make sure the scaling worked.**

```python
dff=pd.DataFrame(sfeatures,columns=df.columns[:-1])
dff.head()
```

# Train Test Split

**Use train_test_split to split your data into a training set and a testing set.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(dff,df['TARGET CLASS'],test_size = 0.30)#, random_state=1)
```

# Using KNN

**Import KNeighborsClassifier from scikit learn.**

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Create a KNN model instance with n_neighbors=1**

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

**Fit this KNN model to the training data.**

```python
knn.fit(X_train,y_train)
```

# Predictions and Evaluations
Let's evaluate our KNN model!

**Use the predict method to predict values using your KNN model and X_test.**

```python
predictions = knn.predict(X_test)
```

**Create a confusion matrix and classification report.**

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

# Choosing a K Value
Let's go ahead and use the elbow method to pick a good K Value!

**Create a for loop that trains various KNN models with different k values, then keep track of the error_rate for each of these models with a list. Refer to the lecture if you are confused on this step.**

```python
import numpy as np
err=[]
for i in range(1,40):
    kn=KNeighborsClassifier(n_neighbors=i)
    kn.fit(X_train,y_train)
    p=kn.predict(X_test)
    err.append(np.mean(p != y_test))
```

**Now create the following plot using the information from your for loop.**

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,40),err,marker='o',markerfacecolor='red')
plt.xlim(0,40)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

It can be seen that least error is in between K=17 to K=21. For a clear understanding of what is the exact value of K at which error rate is minimum, let us draw a subplot of the above graph (i.e., zoomed in version of the same graph).

```python
plt.figure(figsize=(7,9))
plt.plot(range(1,40),err,marker='o',markerfacecolor='red')
plt.xlim(16,21)
plt.ylim(0.15,0.16)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

So, it is clear that for both K=18, we get minimum error of 0.156, therefore, using setting number of neighbours as 18.

## Retrain with new K Value

**Retrain your model with the best K value (up to you to decide what you want) and re-do the classification report and the confusion matrix.**

```python
knn=KNeighborsClassifier(n_neighbors=18)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print("WITH K=18")
print("\n------------------------------------------------------------------------\n")
print(confusion_matrix(y_test,pred))
print("\n------------------------------------------------------------------------\n")
print(classification_report(y_test,pred))
```
